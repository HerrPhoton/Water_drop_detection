На вход поступает черно-белая маска, сделанная в Paint, преобразуем ее в бинарную numpy матрицу True False и сохраняем в эту же папку.

Оригинал


Маска


```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
f = []
for filename in os.listdir():
    if 'mask' in filename and 'npy' not in filename:
        img = cv2.imread(filename)
        masked_img = img.copy()
        masked_img[img > 130] = 1
        masked_img[img <= 130] = 0
        print((masked_img != [0,0,0]).any(2))
        np.save(filename, np.array((masked_img != [0,0,0]).any(2)))
        np.load(filename+'.npy')
```
